```python
#Codes by ask9 https://www.kaggle.com/arbazkhan971/indians-vs-world/notebook

def random_colors(number_of_colors):
    color = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
                 for i in range(number_of_colors)]
    return color
```

```python
from pywaffle import Waffle
import random

gender = df["Gender"].value_counts()

fig = plt.figure(
    FigureClass=Waffle,
    rows=5,
    columns=20,
    values=gender,
    colors = random_colors(len(gender)),
    title={'label': 'Gender at 39 years old Distribution', 'loc': 'left'},
    labels=["{}({})".format(a, b) for a, b in zip(gender.index, gender) ],
    legend={'loc': 'lower left', 'bbox_to_anchor': (0, -0.4), 'ncol': len(df.groupby("Age")), 'framealpha': 0},
    font_size=3,
    icons = 'child',
    figsize=(120, 50),  
    icon_legend=True
)
```
